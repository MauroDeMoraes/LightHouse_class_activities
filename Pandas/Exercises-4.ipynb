```python
import pandas as pd
```

### Download the data and load it to Pandas. 

You can find them [here](https://drive.google.com/file/d/1NY6cmF9Shjw-dD7BD6bNmfcIVz-kQcFR/view?usp=sharing).

```python
cast = pd.read_csv('data/cast.csv', index_col=None)
cast.head()
```

```python
release_dates = pd.read_csv('data/release_dates.csv', index_col=None,
                                      parse_dates=['date'], infer_datetime_format=True)
release_dates.head()
```

### Count the number of movies with "Christmas" in their title for each month, that is released are the USA.

```python
christmas_usa = release_dates.loc[(release_dates['country'] == 'USA') & (release_dates['title'].str.contains("Christmas"))]\
                    .groupby(release_dates['date'].dt.month).title.count().rename_axis('Month').to_frame('Total_movies')
```

```python
christmas_usa
```

### Count the number of movies with "Christmas" in their title for each month, that are released in Canada.

```python
christmas_canada = release_dates.loc[(release_dates['country'] == 'Canada') & (release_dates['title'].str.contains("Christmas"))]\
                    .groupby(release_dates['date'].dt.month).title.count().rename_axis('Month').to_frame('Total_movies')
```

```python
christmas_canada
```

### Count the number of movies in which the titles start with "The Hobbit" for each month, that are released in the USA.

```python
start_the_hobbit_usa = release_dates.loc[(release_dates['country'] == 'USA') & (release_dates['title'].str.startswith('The Hobbit'))]\
                    .groupby(release_dates['date'].dt.month).title.count().rename_axis('Month').to_frame('Total_movies')
```

```python
start_the_hobbit_usa
```

### Count the number of movies with "Romance" in their title for **each day of the week**, that are released in the USA.

```python
romance_usa = release_dates.loc[(release_dates['country'] == 'USA') & (release_dates['title'].str.contains('Romance'))]\
                    .groupby(release_dates['date'].dt.weekday).title.count().rename_axis('Day').to_frame('Total_movies')
```

```python
romance_usa
```

### Count the number of movies with "Action" in their title for **each day of the week**, that are released in the USA.

```python
action_usa = release_dates.loc[(release_dates['country'] == 'USA') & (release_dates['title'].str.contains('Action'))]\
                    .groupby(release_dates['date'].dt.weekday).title.count().rename_axis('Day').to_frame('Total_movies')
```

### On which date was each Judi Dench movie from the 1990s released in the USA?

```python
movie_judi = cast.loc[(cast['name'] == "Judi Dench") & (cast['year'].between(1990,1999))]
movie_released_usa = release_dates.loc[release_dates.country == 'USA']
movie_judi_realse = pd.merge(movie_judi, movie_released_usa, on = 'title' )
```

```python
movie_judi_realse
```

### In which months do films with the actress Judi Dench tend to be released in the USA?

```python
month_release_judi = movie_judi_realse.groupby(movie_judi_realse['date'].dt.month).date.count()\
                    .rename_axis('Month').to_frame('Total movies')
```

```python
month_release_judi
```

### In which months do films with the actor Tom Cruise tend to be released in the USA?

```python
movie_tom = cast.loc[(cast['name'] == "Tom Cruise") & (cast['year'].between(1990,1999))]
movie_released_usa = release_dates.loc[release_dates.country == 'USA']
movie_tom_release = pd.merge(movie_tom, movie_released_usa, on = 'title' )
```

```python
month_release_tom = movie_tom_release.groupby(movie_tom_release['date'].dt.month).date.count()\
                    .rename_axis('Month').to_frame('Total movies')
```
