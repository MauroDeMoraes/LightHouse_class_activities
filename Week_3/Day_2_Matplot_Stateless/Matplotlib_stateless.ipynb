# Matplotlib: Stateless Approach

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Line Plot

```python
#200 values from the interval <0,100>, equidistantly divided
x = np.linspace(0,100,200)
y = np.sin(x)

# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize= (10,6))

# plot the line
ax.plot(x,y,'red')
plt.show()
```

The difference between the stateful and the stateless approaches lies the line of code in which we created the figure and the axes. Because we didn't specify 'nrows' and 'ncols' of the `subplots()` function it returned only one 'axes'. `Axes` object (default settings is `nrows = 1` and `ncols = 1`). Instead of using plt we will use the axes. Axes object to manipulate the plot

### Scatter Plot


```python
#200 random values from the interval <0,100>
x = 10*np.random.rand(200,1)

# 200 random values from the interval <0,15>
y = 15*np.random.rand(200,1)

# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize = (10,6))

# Scatter plot
ax.scatter(x,y)
plt.show()
```

### Histogram

```python
# 200 random values from the interval <0,15>
y = 15*np.random.rand(200,1)

# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize = (10,6))

# histogram
ax.hist(y,bins=20)
plt.show()
```

### Graphs on Common Axes 

```python
# 200 values from the interval <0,100>, equidistantly divides
x = np.linspace(0,100,200)

# sin(x) values
y1 = np.sin(x)

# sin(x)*cos(x) values
y2 = (np.sin(x))*(np.cos(x))

# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize = (10,6))

# sin(x) plot
ax.plot(x,y1, 'red')

# sin(x)*cos(x) plot
ax.plot(x,y,'blue')

plt.show()
```

### Subplots

```python
# create the figure and axes, set the figure size
fig, axes = plt.subplots(nrows = 2, ncols=1, figsize = (10,6))

# the first plot
axes[0].plot(x,y1,'red')

# the first plot title
axes[0].set_title('sin(x)')

# the second plor
axes[1].plot(x,y2,'blue')

# the second title
axes[1].set_title('sin(x)*cos(x)')

# automatically adjust the subplot parameters to give specified padding
fig.tight_layout()
plt.show()
```

In this case, we specified the **nrows** and **ncols** parameters of the `subplots()` function and created two subplots. Axes of these subplots are stored in the **axes** variable (which is a list of **axes.Axes** objects). We can access **axes.Axe**s objects in a list the same way as any other data types stored in a list.


### Legends

```python
import pandas as pd

# import sklearn datasets
from sklearn import datasets
```

```python
# load the iris dataset
iris = datasets.load_iris()

# create a dataframe
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# create a target
iris_df['target'] = iris.target

# map target values to target name
iris_df['target_name']=iris_df.target.map(
    {0: 'setosa',
    1: 'versicolor',
    2: 'virginica'}
)
```

```python
iris_df.head()
```

Now, let's plot a scatter plot of sepal length and sepal width for each type of Iris flower (sepal is the part of a flower which is usually green):

```python
# Iris setosa
setosa = iris_df[iris_df.target_name == 'setosa']

# Iris versicolor
versicolor = iris_df[iris_df.target_name == 'versicolor']

# Iris virginica
virginica = iris_df[iris_df.target_name == 'virginica']

# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize=(10,6))

# scatter the plot of setosa
ax.scatter(setosa['sepal length (cm)'],setosa['sepal width (cm)'],
            marker ='o', color = 'red', label = 'setosa')


# scatter the plot of versicolor
ax.scatter(versicolor['sepal length (cm)'],versicolor['sepal width (cm)'],
            marker ='o', color = 'green', label = 'versicolor')


# scatter the plot of virginica
ax.scatter(virginica['sepal length (cm)'],virginica['sepal width (cm)'],
            marker ='o', color = 'blue', label = 'virginica')

# set the legend
ax.legend(loc='upper right')

# set the title
ax.set_title('Iris flower')

# set the x-label
ax.set_xlabel('sepal length (cm)')

# set the y-label
ax.set_ylabel('sepal width (cm)')
plt.show()
```

### Annotations


```python
# create the figure and anxes, set the figure size
fig, ax = plt.subplots(figsize=(10,6))

# scatter the plot of setosa
ax.scatter(setosa['sepal length (cm)'],setosa['sepal width (cm)'],
            marker ='o', color = 'red', label = 'setosa')


# scatter the plot of versicolor
ax.scatter(versicolor['sepal length (cm)'],versicolor['sepal width (cm)'],
            marker ='o', color = 'green', label = 'versicolor')


# scatter the plot of virginica
ax.scatter(virginica['sepal length (cm)'],virginica['sepal width (cm)'],
            marker ='o', color = 'blue', label = 'virginica')


#### New lines of codes
ax.annotate('setosa', xy = (5.0, 3.5),
            xytext = (4.25, 4.0), arrowprops={'color': 'red'})
ax.annotate('versicolor', xy = (7.2, 3.6),
           xytext = (6.5, 4.0), arrowprops={'color':'red'})
ax.annotate('virginica', xy =(5.05,1.95),
             xytext = (5.5,1.75), arrowprops={'color':'red'})
###########

# set the legend
ax.legend(loc='upper right')

# set the title
ax.set_title('Iris flower')

# set the x-label
ax.set_xlabel('sepal length (cm)')

# set the y-label
ax.set_ylabel('sepal width (cm)')

# set the x-range 
ax.set_xlim(4,8)

# set the y-range
ax.set_ylim(1.5,4.7)
plt.show()
```
