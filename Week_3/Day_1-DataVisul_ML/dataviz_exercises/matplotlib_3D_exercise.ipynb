```python
# import mplot3d
from mpl_toolkits import mplot3d

# import numpy
import numpy as np

# import matplotlib
import matplotlib.pyplot as plt
```

* Create wireframe of hyperbolic paraboloid on the domain <-10,10>X<-10,10>
* hyperbolic paraboloid equation : z = x*x - y*y

```python
# wireframe
def f(x,y):
    return ((x**2)-(y**2))

x = np.linspace(-10, 10, 500)
y = np.linspace(-10, 10, 500)

# a Cartesian product
X,Y = np.meshgrid(x, y)

Z = f(X,Y)

#create the figure and axes
fig, ax = plt.subplots(figsize=(12.8, 9.6))

# set 3D projection
ax = plt.axes(projection='3d')

# a wireframe
ax.plot_wireframe(X,Y,Z, color='blue')
plt.show()
```

* Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>
* eliptical paraboloid : z = x*x + y*y

```python
# surface plot
def f(x,y):
    return ((x**2)+(y**2))

x = np.linspace(-10, 10, 500)
y = np.linspace(-10, 10, 500)

# a Cartesian product
X,Y = np.meshgrid(x, y)

Z = f(X,Y)

#create the figure and axes
fig, ax = plt.subplots(figsize=(12.8, 9.6))

# set 3D projection
ax = plt.axes(projection='3d')

# a surface
ax.plot_surface(X,Y,Z,color='red')
plt.show()
```

* Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>
* Create also contour projection on the plane z=0 (HINT: use parameter 'zdir' of contour plot)
* eliptical paraboloid : z = x*x + y*y

```python
# surface plot
def f(x,y):
    return ((x**2)+(y**2))

x = np.linspace(-10, 10, 500)
y = np.linspace(-10, 10, 500)

# a Cartesian product
X,Y = np.meshgrid(x, y)

Z = f(X,Y)

#create the figure and axes
fig, ax = plt.subplots(figsize=(12.8, 9.6))

# set 3D projection
ax = plt.axes(projection='3d')

# a surface
ax.plot_surface(X,Y,Z, color='green')


# a surface
ax.contour(X,Y,Z, 40, zdir='Z', cmap = 'jet', offset = 0)
plt.show()




```
