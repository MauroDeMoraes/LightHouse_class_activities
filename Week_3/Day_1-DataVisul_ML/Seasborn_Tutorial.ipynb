#### replot()

This is a figure-level-function that makes use of two other axes functions for Visualizing Statistical Relationships which are:
* `scatterplot()`
* `lineplot()`


```python
f = sns.load_dataset("flights")
sns.relplot(x = "passengers", y ="month", data = f)
```

As you can see, the points are plotted in 2-dimensions. However, you can add another dimension using the `hue` semantic. Letâ€™s take a look at an example of the same:

```python
f = sns.load_dataset("flights") 
sns.relplot(x="passengers", y="month", hue="year", data=f);
```

Let me just show how you can change the color in the following example:

```python
sns.set(style="darkgrid")
f = sns.load_dataset("flights")
sns.relplot(x="passengers", y="month", hue="year",palette="ch:r=-.5,l=.75", data=f);
```

## Plotting with Categorical Data:

This approach comes into the picture when our main variable is further divided into discrete groups (categorical). This can be achieved using the `catplot()` function.

#### catplot():

This is a figure-level-function like relplot(). It can be characterized by three families of axes level functions namely:
1. Scatterplots â€” These include `stripplot(`), `swarmplot()`
2. Distribution Plots â€” which are `boxplot()`, `violinplot()`, `boxenplot()`
3. Estimateplots â€” namely `pointplot()`, `barplot()`, `countplot()`

```python
sns.set(style='ticks', color_codes = True)
a = sns.load_dataset("tips")
sns.catplot(x='day', y='total_bill', data=a)
```

As you can see, in the above example I have not set the â€˜kindâ€™ parameter. Therefore it has returned the graph as the default scatterplot. You can specify any of the axes level function to change the graph as need be.

```python
sns.set(style='ticks', color_codes=True)
a = sns.load_dataset("tips")
sns.catplot(x='day', y='total_bill', kind='violin', data=a)
```

### Multi-Plot Grids:

Python Seaborn allows you to plot multiple grids side-by-side. These are basically plots or graphs that are plotted using the same scale and axes to aid comparison between them. This, in turn, helps the programmer to differentiate quickly between the plots and obtain large amounts of information.

Consider the following example of `facetgrid()` function to plot these graphs.

```python
sns.set(style='darkgrid')
a = sns.load_dataset('iris')
b = sns.FacetGrid(a, col='species')
b.map(plt.hist, "sepal_length")
```

You can also plot using PairGrid function when you have a pair of variables to compare.

```python
sns.set(style="ticks")
a = sns.load_dataset("flights")
b = sns.PairGrid(a)
b.map(plt.scatter)
```

As you can see, the above output clearly compares between the year and the number of passengers in different ways.

## Plot-Aesthetics

This segment of Python Seaborn tutorial deals with making our plots more attractive and delightful.

#### Python Seaborn Figure-Aesthetics:

The first function that I shall be discussing is `set()`. I have been using the â€˜styleâ€™ parameter of this function before. This parameter basically deals with seaborn themes. At present, there are five of them available namely darkgrid, ticks, whitegrid, white and dark.

```python
sns.set(style='white', color_codes=True)
a = sns.load_dataset('tips')
sns.boxplot(x='day', y="total_bill", data=a)
```

In the above output, you can notice the theme is changed to white. You can explore them further using the other themes as well. If you notice in the previous output, there are axes present all around the graph. However, this is also customizable using the `despine()` function. Look at the example below.

```python
sns.set(style='white', color_codes=True)
a = sns.load_dataset('tips')
sns.boxplot(x='day', y='total_bill', data=a)
sns.despine(offset=10, trim=True)
```

# Regression Line With a Confidence Interval

A regression line with confidence interval can be handy, for example, when we want to plot the dependence of a tip based on the total bill. It can be done by computing the regression line and the confidence interval with some library (sklearn, statsmodels, SciPy) and then plot the result using Matplotlib. But let's see how easily it can be done with Seaborn.



```python
# a regression line woth a confidence interval plot
sns.lmplot(x='total_bill', y='tip', data = tips)
plt.show()
```

Now if we want to have two regression lines in one chart â€“ one for female tips and one for male tips â€“ we can modify the code above and use the `hue` parameter.

```python
sns.lmplot(x='total_bill', y='tip', data=tips,hue='sex')
plt.show()
```

On the other hand, we could also have one plot for each of the regression lines and write only one line of code.

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
plt.show()
```

## Residual Plot

Another plot that is related to the regression line is a residual plot. We don't need to compute residuals manually and then plot the results. Instead, we can use a built-in function

```python
sns.residplot(x='total_bill', y='tip', data= tips)
plt.show()
```

### Strip Plot

```python
sns.stripplot(x='day',y ='tip',data = tips, size =5, jitter = True)

# overwrite the original y-label from 'tip' to 'tip ($)'
plt.ylabel('tip ($)')
plt.show()
```

### Swarm Plot


```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')

# overwrite the original y-label from 'tip' tp 'tip($)'
plt.ylabel('tip (#)')
plt.show
```

### Boxplot + Violin Plot

```python
# create a grid with 1 row 2 columns
plt.subplot(1,2,1)

# first plot
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2)

# The second plot
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.tight_layout()
plt.show
```

### Joint Plot

Joint plot with density estimates:

### Pair Plotz

A pair plot with the `hue` parameter
